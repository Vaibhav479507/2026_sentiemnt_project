```python
import joblib
import re
```

```python
def mycleaning(doc):
    return re.sub("[^a-zA-Z]"," ",doc).lower()
model=joblib.load('sentiment_model.pkl')

sample=input('enter review')
pred=model.predict([sample])
prob=model.predict_proba([sample])
if pred[0]==0:
    print("Neg")
    print(f"Confidence Score : {prob[0][0]:.2f}")
else:
    print("Pos")
    print(f"Confidence Score : {prob[0][1]:.2f}")
```
